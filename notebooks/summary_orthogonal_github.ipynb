# Structure of NG integrals

The structure of the $12$ integrals is the following. We define the primordial bispectrum as e.g. in https://arxiv.org/pdf/1906.08697.pdf, Eq. (2.4). This corresponds to the Planck convention, and to the TASI lectures of Baumann (which is what I had used in the first place). **For more details, see the notes on orthogonal NG, which have all the definitions and factors.**

The bispectrum is scale-invariant, and takes the form 
\begin{equation*}
B(k_1,k_2,k_3) \propto \frac{{\cal S}(k_1,k_2,k_3)}{k_1^2k_2^2k_3^2}\,\,,
\end{equation*}
where the shape function is dimensionless. Then, it is natural to compute the different contributions as
\begin{equation*}
\mu^{2n}{\cal P}(k)\int_\mathbf{q} {\cal S}'(k,q,|{\mathbf{k}-\mathbf{q}}|) {\cal P}(q) {\cal P}(|{\mathbf{k}-\mathbf{q}}|)\,\,,
\end{equation*}
where ${\cal S}'$ is some dimensionless shape built from ${\cal S}$ and the kernels of the bias expansion ($F_2$, $G_2$, ${\cal G}_2$, and so on), and
\begin{equation*}
{\cal P}(k) = \frac{{\cal M}(k)}{k^2}\,\,, 
\end{equation*}
where we recall that (see e.g.~the bias review) 
\begin{equation*}
{\cal M}(k) = \frac{2}{3}\frac{k^2T(k)D_{\rm md}(\tau)}{\Omega_{m,0}H^2_0}\,\,. 
\end{equation*}

But let's be extra-precise with the factors, since this helps keeping track of the modifications to the code. We have that the one-loop contributions are of the form
\begin{equation*}
\bigg({\frac{1944}{625}}f_{\rm NL}\pi^4A^2_{\rm s}\bigg){\cal P}(k)\mu^{2n}\int(2\times\text{dimensionless function ${\cal S}'$ $\to$ FFT matrix}) {\cal P}(q)  {\cal P}(|{\mathbf{k}-\mathbf{q}}|)\,\,,
\end{equation*}
where I emphasize that the matrices already account for the overall $2$ that comes from the permutations!!! 

Then, we recall that
\begin{equation*}
P_\delta(k) = \frac{9}{25}P_\zeta(k){\cal M}^2(k)\,\,.
\end{equation*}
Now, my function *nonlinear\_pt\_pPRIMk\_l* in "nonlinear_pt.c" extracts
\begin{equation*}
\frac{k^4\times P_\zeta(k)}{A_{\rm s}}\,\,. 
\end{equation*}
Then, in the code I define a ${\cal T}$ (appearing as *Tbin*, *Tnw*, *Tw* in the code, mimicking the original CLASS-PT), which is 
\begin{equation*}
{\cal T}(k) = \sqrt{A_{\rm s}}\,{\cal P}(k)\,\,. 
\end{equation*}
Its wiggly and non-wiggly parts are extracted from the linear power spectrum that CLASS-PT already gets, and the output of my function *nonlinear\_pt\_pPRIMk\_l*. The wiggly part is extracted via perturbation theory in the wiggles up to (uselessly) high order. I had checked that even stopping at first order was enough for all practical values of $\Omega_b$. 

Then, if we work with ${\cal T}$ instead of ${\cal P}$, we see that the overall factor needed is 
\begin{equation*}
\bigg({\frac{1944}{625}}\pi^4f_{\rm NL}\sqrt{A_{\rm s}}\bigg)\,\,,
\end{equation*}
where we notice the nice combination $f_{\rm NL}\sqrt{A_{\rm s}}$ appearing.

# Counterterms, scale-invariant bispectra, etc.

An important point to emphasize is the following. Now, we are using scale-invariant bispectra, which is the same as Planck does. I.e., if I were to consider the local shape, I do *not* have that the bispectrum is given by $P_\zeta(k_1)P_\zeta(k_2) + \dots$, where $P_\zeta(k)$ is tilted. 

This has the consequence that the counterterm for equilateral (and orthogonal) is *not* of the form $\nabla^2\phi$, which differs from $b_1$ only due to the transfer function. It is actually of the form

$$P_{\rm counter}(k) \sim \frac{P(k)}{T(k)}\times\bigg(\frac{k}{k_\star}\bigg)^{1-n_{\rm s}}\,\,.$$

So it differs from $b_1$ both due to the transfer function, *and* the tilt. Fixing the tilt can be tricky since we do not know a non-scale-invariant orthogonal template.

**We checked in the paper that for BOSS it is useless, it does not change constraints...**
