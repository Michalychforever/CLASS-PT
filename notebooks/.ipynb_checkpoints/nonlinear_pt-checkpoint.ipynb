Ok. So now we go to the convenience functions...

$$pk\_m0 = 0\text{pole_vv}$$
$$pk\_m2 = 2\text{pole_vv}$$
$$pk\_m4 = 4\text{pole_vv}$$

$$pk\_g0 = 0\text{pole_vv} + b1\,0\text{pole_vd} + b1^2\,0\text{pole_dd} + b1\,\frac{b2}{2}\,0\text{pole_b1b2} + \frac{b2}{2}\,0\text{pole_b2} + b1 \, {bG2}\,0\text{pole_b1bG2} + {bG2}\,0\text{pole_bG2}$$
$$pk\_g2 = 2\text{pole_vv} + b1\,2\text{pole_vd} + b1^2\,2\text{pole_dd} + b1\,\frac{b2}{2}\,2\text{pole_b1b2} + \frac{b2}{2}\,2\text{pole_b2} + b1 \, {bG2}\,2\text{pole_b1bG2} + {bG2}\,2\text{pole_bG2}$$
$$pk\_g4 = 4\text{pole_vv} + b1\,4\text{pole_vd} + b1^2\,4\text{pole_dd} + b1\,\frac{b2}{2}\,4\text{pole_b1b2} + \frac{b2}{2}\,4\text{pole_b2} + b1 \, {bG2}\,4\text{pole_b1bG2} + {bG2}\,4\text{pole_bG2}$$

*Notice that I do not need to put a factor of 2 in front of the vd term!!!* The reason is that I have extracted exactly the object that multiplies b1. For Misha and Oliver it is the same: just see their utility functions... Then, notice that I had extracted $b_2/2$, *not* $b_2$ only... All operations for extracting multipoles and stuff, multiplying by $Z_1$, and so on, are linear, so this still holds...
