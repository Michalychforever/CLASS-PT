```python
z_pk = 0.61

common_settings = {'ln10^{10}A_s':3.044,
                   'T_cmb':2.726,
                   'n_s':0.9649,
                   'z_reio':10.,
                   'k_pivot':0.05,
                   'omega_b':0.02237,
                   'omega_cdm':0.12,
                   'h':0.6736,
                   'YHe':0.25,
                   'N_eff':3.046,
                   'N_ncdm':0,
                   #'Omega_ncdm':0., #--> if 'N_ncdm':0 this is not read, which of course makes sense...
                   'z_pk':z_pk}
```

```python
M128 = Class()
M128.set(common_settings)

M128.set({'output':'mPk',
        'non linear':'PT',
        'IR resummation':'No',
        'Bias tracers':'Yes',
        'cb':'Yes',
        'RSD':'No',
        'FFTLog mode':'FAST'})#,
        #'AP':'Yes',
        #'Omfid':'0.31'})

M128.compute()
```

```python
M256 = Class()
M256.set(common_settings)

M256.set({'output':'mPk',
        'non linear':'PT',
        'IR resummation':'No',
        'Bias tracers':'Yes',
        'cb':'Yes',
        'RSD':'No',
        'FFTLog mode':'NORMAL'})#,
        #'AP':'Yes',
        #'Omfid':'0.31'})

M256.compute()
```

```python
M512 = Class()
M512.set(common_settings)

M512.set({'output':'mPk',
        'non linear':'PT',
        'IR resummation':'No',
        'Bias tracers':'Yes',
        'cb':'Yes',
        'RSD':'No',
        'FFTLog mode':'NORMAL'})#,
        #'AP':'Yes',
        #'Omfid':'0.31'})

M512.compute()
```

```python
font={'size':20,'family':'STIXGeneral'}
axislabelfontsize='large'
matplotlib.rc('font',**font)
matplotlib.mathtext.rcParams['legend.fontsize']='small'
plt.rcParams["figure.figsize"]=[8.0,6.0]
```

```python
numberdict={
            "matter":48,
            "b2":49,
            "bG2":50}
```

```python
h=0.6736

kvec=np.logspace(-3,np.log10(0.3),1000) #--> array of k in h/Mpc (NOTA BENE!!!)
twopi = 2.*math.pi
khvec = kvec*h #--> array of k in 1/Mpc (NOTA BENE!!!)

M128_mult = M128.get_pk_mult(khvec,z_pk,len(khvec))
M256_mult = M256.get_pk_mult(khvec,z_pk,len(khvec))
M512_mult = M512.get_pk_mult(khvec,z_pk,len(khvec))

classpts128={}
classpts256={}
classpts512={}

for el in numberdict:
    classpts128[el]=M128_mult[numberdict[el]]*h**3.
    classpts256[el]=M256_mult[numberdict[el]]*h**3.
    classpts512[el]=M512_mult[numberdict[el]]*h**3.
```

```python
#--> matter

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(classpts128['matter'])),color='blue',linestyle='-',label='128')
ax_Pk.loglog(kvec,np.array(np.abs(classpts256['matter'])),color='red',linestyle='-',label='256')
ax_Pk.loglog(kvec,np.array(np.abs(classpts512['matter'])),color='green',linestyle='-',label='512')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.0001,100])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("matter")
fig_Pk.tight_layout()

#--> b2

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(classpts128['b2'])),color='blue',linestyle='-',label='128')
ax_Pk.loglog(kvec,np.array(np.abs(classpts256['b2'])),color='red',linestyle='-',label='256')
ax_Pk.loglog(kvec,np.array(np.abs(classpts512['b2'])),color='green',linestyle='-',label='512')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.001,1000])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("b2")
fig_Pk.tight_layout()

#--> bG2

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(classpts128['bG2'])),color='blue',linestyle='-',label='128')
ax_Pk.loglog(kvec,np.array(np.abs(classpts256['bG2'])),color='red',linestyle='-',label='256')
ax_Pk.loglog(kvec,np.array(np.abs(classpts512['bG2'])),color='green',linestyle='-',label='512')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.001,100])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("bG2")
fig_Pk.tight_layout()
```
