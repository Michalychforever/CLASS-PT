# Structure of NG integrals

The structure of the $12$ integrals is the following. We define the primordial bispectrum as e.g. in https://arxiv.org/pdf/1906.08697.pdf, Eq. (2.4). This corresponds to the Planck convention, and to the TASI lectures of Baumann (which is what I had used in the first place).

The bispectrum is scale-invariant, and takes the form 
\begin{equation*}
B(k_1,k_2,k_3) \propto \frac{{\cal S}(k_1,k_2,k_3)}{k_1^2k_2^2k_3^2}\,\,,
\end{equation*}
where the shape function is dimensionless. Then, it is natural to compute the different contributions as
\begin{equation*}
\mu^{2n}{\cal P}(k)\int_\mathbf{q} {\cal S}'(k,q,|{\mathbf{k}-\mathbf{q}}|) {\cal P}(q) {\cal P}(|{\mathbf{k}-\mathbf{q}}|)\,\,,
\end{equation*}
where ${\cal S}'$ is some dimensionless shape built from ${\cal S}$ and the kernels of the bias expansion ($F_2$, $G_2$, ${\cal G}_2$, and so on), and
\begin{equation*}
{\cal P}(k) = \frac{{\cal M}(k)}{k^2}\,\,, 
\end{equation*}
where we recall that (see e.g.~the bias review) 
\begin{equation*}
{\cal M}(k) = \frac{2}{3}\frac{k^2T(k)D_{\rm md}(\tau)}{\Omega_{m,0}H^2_0}\,\,. 
\end{equation*}

But let's be extra-precise with the factors, since this helps keeping track of the modifications to the code. We have that the one-loop contributions are of the form
\begin{equation*}
\bigg({\frac{1944}{625}}f_{\rm NL}\pi^4A^2_{\rm s}\bigg){\cal P}(k)\mu^{2n}\int(2\times\text{dimensionless function ${\cal S}'$ $\to$ FFT matrix}) {\cal P}(q)  {\cal P}(|{\mathbf{k}-\mathbf{q}}|)\,\,,
\end{equation*}
where I emphasize that the matrices already account for the overall $2$ that comes from the permutations!!! 

Then, we recall that
\begin{equation*}
P_\delta(k) = \frac{9}{25}P_\zeta(k){\cal M}^2(k)\,\,.
\end{equation*}
Now, my function *nonlinear\_pt\_pPRIMk\_l* in "nonlinear_pt.c" extracts
\begin{equation*}
\frac{k^4\times P_\zeta(k)}{A_{\rm s}}\,\,. 
\end{equation*}
Then, in the code I define a ${\cal T}$ (appearing as *Tbin*, *Tnw*, *Tw* in the code, mimicking the original CLASS-PT), which is 
\begin{equation*}
{\cal T}(k) = \sqrt{A_{\rm s}}\,{\cal P}(k)\,\,. 
\end{equation*}
Its wiggly and non-wiggly parts are extracted from the linear power spectrum that CLASS-PT already gets, and the output of my function *nonlinear\_pt\_pPRIMk\_l*. The wiggly part is extracted via perturbation theory in the wiggles up to (uselessly) high order. I had checked that even stopping at first order was enough for all practical values of $\Omega_b$. 

Then, if we work with ${\cal T}$ instead of ${\cal P}$, we see that the overall factor needed is 
\begin{equation*}
\bigg({\frac{1944}{625}}\pi^4f_{\rm NL}\sqrt{A_{\rm s}}\bigg)\,\,,
\end{equation*}
where we notice the nice combination $f_{\rm NL}\sqrt{A_{\rm s}}$ appearing.

# Summary of overall factors

**In summary** $\to$ the outputs of CLASS-PT below need to be multiplied by 
\begin{equation*}
\bigg({\frac{1944}{625}}\pi^4f_{\rm NL}\sqrt{A_{\rm s}}\bigg)\,\,.
\end{equation*}

Also, I emphasize that all the contributions where $b_2$ appears get multiplied by $b_2/2$. This is the same convention as the old CLASS-PT. Also, notice that $b_2$ and $b_{{\cal G}_2}$ always appear linearly.

# Structure of IR resummation

The IR resummation was implemented in the following way. We see that the $12$ terms are essentially a mixed $13$ and $22$ form. The IR resummation was done, then, in a mixed way. For the external ${\cal T}$, the IR resummation was done including the $\mu$-dependent damping, as in the $P_{13}$-like diagrams. The internal piece, instead, used the same structure of IR resummation of the $22$-like FFTlog matrix multiplications. For details, see the .tex file in the source folder where I summarize what the original CLASS-PT does...

# Powers of Hubble

Nothing much to say here. The code is the same as the old CLASS-PT. It eats $k$ in $1/{\rm Mpc}$, and the outputs are in ${\rm Mpc}^3$.

This is different for the "utility functions" of Oliver, that have now been augmented to include, e.g., the combination of all the terms forming the galaxy-galaxy "$12$" multipoles. They eat $k$ in $1/{\rm Mpc}$, but already spit out the power spectra in ${\rm Mpc}^3/h^3$.

The examples below are self-explanatory.

# Counterterms, scale-invariant bispectra, etc.

An important point to emphasize is the following. Now, we are using scale-invariant bispectra, which is the same as Planck does. I.e., if I were to consider the local shape, I do *not* have that the bispectrum is given by $P_\zeta(k_1)P_\zeta(k_2) + \dots$, where $P_\zeta(k)$ is tilted. 

This has the consequence that the counterterm for equilateral is *not* of the form $\nabla^2\phi$, which differs from $b_1$ only due to the transfer function. It is actually of the form

$$P_{\rm counter}(k) \sim \frac{P(k)}{T(k)}\times\bigg(\frac{k}{k_\star}\bigg)^{1-n_{\rm s}}\,\,.$$

So it differs from $b_1$ both due to the transfer function, *and* the tilt.

**In any case we decided to not add this, since for BOSS it should be completely useless...**

# Removal of "useless" matrices

While writing this code, an issue arised. Computing the matrices with recursion relations in C was complicated. Then, I resorted to just compute them outside and import them. There are many of these matrices, and since they are read every time from disk, there could be a significant hit in performance.

In reality the hit is not so bad. However, since in the real world we only care about the case where IR resummation and AP are turned on, there is no need to import all the matrices. I then constructed a version of "nonlinear_pt.c" where I import only the matrix for the matter-matter in real space, and the matrices for the $\mu$ powers of matter, and the matrices for the multipoles of $b_2$ and $b_{{\cal G}_2}$. These are the only ones needed. (Technically, the matter-matter in real space is not needed, but I didn't want to modify the code too much. One matrix more is nothing...) Then, the matrices for the real-space bias contributions, and the multipoles pre-IR and pre-AP, are equal to the matter-matter in real space...

So in the final version of the code those results are wrong :)

# Addition of constant for logarithms

Notice that in CLASS-PT one needs to add an overall constant to the power spectra, since they sometime go negative and then interpolation fails since it is done in log-log space. In my case, I add $1000$ to all the terms. I think this is also maybe too much (especially for the terms that are generated only from AP and IR resummation), since now $f_{\rm NL}$ is outside, so I will not have wildly different power spectra if I sample with $\Lambda$CDM parameters close to Planck...

In any case, this is a variable (called "large_for_logs_fNL" or something) in both the C and wrapper, so it can be changed accordingly...

````python
z_pk = 0.61

common_settings = {'ln10^{10}A_s':3.044,
                   'T_cmb':2.726,
                   'n_s':0.9649,
                   'z_reio':10.,
                   'k_pivot':0.05,
                   'omega_b':0.02237,
                   'omega_cdm':0.12,
                   'h':0.6736,
                   'YHe':0.25,
                   'N_eff':3.046,
                   'N_ncdm':0,
                   #'Omega_ncdm':0., #--> if 'N_ncdm':0 this is not read, which of course makes sense...
                   'z_pk':z_pk}  

M = Class()
M.set(common_settings)

M.set({'output':'mPk',
        'non linear':'PT',
        'IR resummation':'Yes',
        'Bias tracers':'Yes',
        'cb':'Yes',
        'RSD':'Yes',
        'AP':'Yes',
        'Omfid':'0.31'})

M.compute()
````

How do these contribute to the galaxy-galaxy multipoles? Let us make the following definitions:

\begin{equation*}
0_{\rm vv} = \text{monopole, no powers of $b_1$}
\end{equation*}
\begin{equation*}
0_{\rm vd} = \text{monopole, one power of $b_1$}
\end{equation*}
\begin{equation*}
0_{\rm dd} = \text{monopole, two powers of $b_1$}
\end{equation*}

\begin{equation*}
2_{\rm vv} = \text{quadrupole, no powers of $b_1$}
\end{equation*}
\begin{equation*}
2_{\rm vd} = \text{quadrupole, one power of $b_1$}
\end{equation*}
\begin{equation*}
