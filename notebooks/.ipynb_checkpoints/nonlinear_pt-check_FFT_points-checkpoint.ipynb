Ok, I tried renaming the matrices and indeed it crashes if it doesn't find them, given the 'FFTLog mode' flag. But I don't get why when I was running my checks in Mathematica, I was always outputting 512 points. The reason is the following: that was the length of what in Mathematica is called "kNplot". This has NOT to be the same as the size of the matrix you use for the FFTlog/the number of coefficients for the PS. Just see the Mathematica notebooks: they loop over kNplot, and for each of them they do the matrix multiplication, which is an operation that involves arrays of a dimension fixed by the "kN" length! Nice... So, when I was comparing with Mathematica, I was using 256 in C (the default!!!), and 512 in Mathematica. Still, the comparison led to a very very small relative error!!!

```python
numberdict={
            "0pole_b1b2":60,
            "0pole_b1bG2":62,
            "0pole_b2":61,
            "0pole_bG2":63,
            "2pole_b1b2":64,
            "2pole_b1bG2":66,
            "2pole_b2":65,
            "2pole_bG2":67,
            "4pole_b1b2":68,
            "4pole_b1bG2":70,
            "4pole_b2":69,
            "4pole_bG2":71,
            "0pole_vv":51,
            "0pole_vd":52,
            "0pole_dd":53,
            "2pole_vv":54,
            "2pole_vd":55,
            "2pole_dd":56,
            "4pole_vv":57,
            "4pole_vd":58,
            "4pole_dd":59}
```

```python
h=0.6736

kvec=np.logspace(-3,np.log10(0.3),1000) #--> array of k in h/Mpc (NOTA BENE!!!)
khvec = kvec*h #--> array of k in 1/Mpc (NOTA BENE!!!)

M128_mult = M128.get_pk_mult(khvec,z_pk,len(khvec))
M256_mult = M256.get_pk_mult(khvec,z_pk,len(khvec))
M512_mult = M512.get_pk_mult(khvec,z_pk,len(khvec))

classpts128={}
classpts256={}
classpts512={}

for el in numberdict:
    classpts128[el]=M128_mult[numberdict[el]]*h**3.
    classpts256[el]=M256_mult[numberdict[el]]*h**3.
    classpts512[el]=M512_mult[numberdict[el]]*h**3.
```
