```python
z_pk = 0.61

common_settings = {'ln10^{10}A_s':3.044,
                   'T_cmb':2.726,
                   'n_s':0.9649,
                   'z_reio':10.,
                   'k_pivot':0.05,
                   'omega_b':0.02237,
                   'omega_cdm':0.12,
                   'h':0.6736,
                   'YHe':0.25,
                   'N_eff':3.046,
                   'N_ncdm':0,
                   #'Omega_ncdm':0., #--> if 'N_ncdm':0 this is not read, which of course makes sense...
                   'z_pk':z_pk}  

M = Class()
M.set(common_settings)

M.set({'output':'mPk',
        'non linear':'PT',
        'IR resummation':'No',
        'Bias tracers':'Yes',
        'cb':'Yes',
        'RSD':'No'})#,
        #'AP':'Yes',
        #'Omfid':'0.31'})

M.compute()
```

```python
font={'size':20,'family':'STIXGeneral'}
axislabelfontsize='large'
matplotlib.rc('font',**font)
matplotlib.mathtext.rcParams['legend.fontsize']='small'
plt.rcParams["figure.figsize"]=[8.0,6.0]
```

```python
importdict={
            "matter":"output_matter_no_ir_resummation.txt",
            "b2":"output_b2_no_ir_resummation.txt",
            "bG2":"output_bG2_no_ir_resummation.txt"}

numberdict={
            "matter":48,
            "b2":49,
            "bG2":50}
```

```python
folder="../comparisons_with_mathematica/"

dataimportdict={}

for el in importdict:
    dataimportdict[el]=np.loadtxt(folder+importdict[el])
```

```python
interpdict={}

for el in importdict:
    interpdict[el]=interpolate.interp1d(dataimportdict[el][:,0],dataimportdict[el][:,1],kind='cubic')
    
#--> NOTICE THAT THESE ARE STILL IN Mpc...
```

```python
h=M.h()

fz=M.scale_independent_growth_factor_f(z_pk) #--> unneeded...

kvec=np.logspace(-3,np.log10(0.3),1000) #--> array of k in h/Mpc (NOTA BENE!!!)
twopi = 2.*math.pi
khvec = kvec*h #--> array of k in 1/Mpc (NOTA BENE!!!)

M_mult = M.get_pk_mult(khvec,z_pk,len(khvec))

mypts={}
classpts={}

for el in importdict:
    mypts[el]=interpdict[el](khvec)*h**3.
    classpts[el]=M_mult[numberdict[el]]*h**3.
```

```python
#--> matter

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(mypts['matter'])),color='blue',linestyle='-',label='C')
ax_Pk.loglog(kvec,np.array(np.abs(classpts['matter'])),color='red',linestyle='-',label='classy')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.0001,100])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("matter")
fig_Pk.tight_layout()

#--> b2

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(mypts['b2'])),color='blue',linestyle='-',label='C')
ax_Pk.loglog(kvec,np.array(np.abs(classpts['b2'])),color='red',linestyle='-',label='classy')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.001,1000])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("b2")
fig_Pk.tight_layout()

#--> bG2

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(mypts['bG2'])),color='blue',linestyle='-',label='C')
ax_Pk.loglog(kvec,np.array(np.abs(classpts['bG2'])),color='red',linestyle='-',label='classy')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.001,100])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("bG2")
fig_Pk.tight_layout()
```

Ok. So now we go to the convenience functions...

$$pk\_mm = \text{matter}$$
$$pk\_gm = b1\,\text{matter} + \frac{b2}{2}\,\text{b2} + bG2\,\text{bG2}$$
$$pk\_gg = b1^2\,\text{matter} + b1\,\frac{b2}{2}\,\text{b2} + b1\,bG2\,\text{bG2}$$

```python
M.initialize_output(khvec,z_pk,len(khvec))

b1=1.2
b2=0.3
bG2=-0.2

pk_mm=M.pk_mm_fNL_real()
pk_gm=M.pk_gm_fNL_real(b1,b2,bG2)
pk_gg=M.pk_gg_fNL_real(b1,b2,bG2)

pk_mmmine=classpts['matter']
pk_gmmine=b1*classpts['matter'] + b2/2.*classpts['b2'] + bG2*classpts['bG2']
pk_ggmine=b1**2.*classpts['matter'] + b1*b2/2.*classpts['b2'] + b1*bG2*classpts['bG2']
```

```python
#--> mm

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(pk_mm)),color='blue',linestyle='-',label='classy')
ax_Pk.loglog(kvec,np.array(np.abs(pk_mmmine)),color='red',linestyle='-',label='classy -- mine')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.0001,100])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("mm")
fig_Pk.tight_layout()

#--> gm

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(pk_gm)),color='blue',linestyle='-',label='C')
ax_Pk.loglog(kvec,np.array(np.abs(pk_gmmine)),color='red',linestyle='-',label='classy')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.001,100])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("gm")
fig_Pk.tight_layout()

#--> gg

fig_Pk,ax_Pk=plt.subplots()

ax_Pk.loglog(kvec,np.array(np.abs(pk_gg)),color='blue',linestyle='-',label='C')
ax_Pk.loglog(kvec,np.array(np.abs(pk_ggmine)),color='red',linestyle='-',label='classy')

ax_Pk.set_xlim([1.e-3,0.3])
ax_Pk.set_ylim([0.001,100])
ax_Pk.set_xlabel(r'$k \,\,\,\, [h\mathrm{Mpc}^{-1}]$')
ax_Pk.set_ylabel(r'$P_{\ell}(k)\,\,\,\, [h^{-1}\mathrm{Mpc}]^{3}$')
ax_Pk.legend(fontsize='16',ncol=1,loc='upper right')
ax_Pk.set_title("gg")
fig_Pk.tight_layout()
```
